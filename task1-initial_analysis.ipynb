## **Transforming the Data**
The output values are 'Attack' or 'Natural'. I have changed them to 1 and 0 respectively

```python
df_cleaned['marker_encoded'] = df_cleaned['marker'].map({'Attack': 1, 'Natural': 0})
df_cleaned.head()
```

```python
final_data = df_cleaned.copy(deep=True)
final_data = final_data.drop('marker',axis=1)
final_data.head()
```

```python
final_data['marker_encoded'].value_counts()
```

```python
shuffled_data = final_data.sample(frac=1, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(shuffled_data, test_size=0.20)  # 80% for training , 20% for testing
```

```python
print(f"train_set shape: {train_set.shape}\n")
print(f"test_set shape: {test_set.shape}\n")
```

```python
category_counts = train_set.iloc[:, -1].value_counts()
print(category_counts)
```

```python
category_counts = test_set.iloc[:, -1].value_counts()
print(category_counts)
```

```python
X_train = train_set.drop('marker_encoded', axis=1)
y_train = train_set['marker_encoded']

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
X_test = test_set.drop('marker_encoded', axis=1)
y_test = test_set['marker_encoded']

print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

## **Checking how normalizing the data helps with acheiving good results**

### Data normalization using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```

 ### Training the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import xgboost as xgb
import matplotlib.pyplot as plt

# Initializing and training the models
logreg_normalized = LogisticRegression(max_iter=3000).fit(X_train_normalized, y_train)
decision_tree_normalized = DecisionTreeClassifier().fit(X_train_normalized, y_train)
random_forest_normalized = RandomForestClassifier().fit(X_train_normalized, y_train)
xgb_model_normalized = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss').fit(X_train_normalized, y_train)

# Generating the predictions
predictions_normalized = {
    'Logistic Regression': logreg_normalized.predict(X_test_normalized),
    'Decision Tree': decision_tree_normalized.predict(X_test_normalized),
    'Random Forest': random_forest_normalized.predict(X_test_normalized),
    'XGBoost': xgb_model_normalized.predict(X_test_normalized)
}

for name, y_pred_normalized in predictions_normalized.items():
    print(f"Classification Report for {name} with Normalized Data:")
    print(classification_report(y_test, y_pred_normalized))
    print("-" * 50)
```

### confusion matrix

```python
fig, axes = plt.subplots(1, len(predictions_normalized), figsize=(20, 5))
fig.suptitle('Confusion Matrices for Normalized Data')
for ax, (name, y_pred) in zip(axes.flatten(), predictions_normalized.items()):
    disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax, cmap='Blues')
    disp.ax_.set_title(name)
plt.tight_layout()
plt.show()
```

### cross validation

```python
from sklearn.model_selection import cross_val_score

models = {
    'Logistic Regression': logreg_normalized,
    'Decision Tree': decision_tree_normalized,
    'Random Forest': random_forest_normalized,
    'XGBoost': xgb_model_normalized
}

scoring_metrics = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

val_score_data = []

for metric in scoring_metrics:
    for name, model in models.items():
        scores = cross_val_score(model, X_train_normalized, y_train, scoring=metric, cv=10, n_jobs=-1)
        val_score_data.append({
            'Model': name,
            'Metric': metric,
            'Mean Score': np.mean(scores),
            'Standard Deviation': np.std(scores)
        })
val_score_df = pd.DataFrame(val_score_data)

val_score_df_pivot = val_score_df.pivot(index='Model', columns='Metric', values=['Mean Score', 'Standard Deviation'])

print(val_score_df_pivot)
```

## **Checking how standardizing the data helps with acheiving good results**

### Data Standardization using StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_standardized = scaler.fit_transform(X_train)
X_test_standardized = scaler.transform(X_test)
```

### Training the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import xgboost as xgb
import matplotlib.pyplot as plt

logreg_standardized = LogisticRegression(max_iter=3000).fit(X_train_standardized, y_train)
decision_tree_standardized = DecisionTreeClassifier().fit(X_train_standardized, y_train)
random_forest_standardized = RandomForestClassifier().fit(X_train_standardized, y_train)
xgb_model_standardized = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss').fit(X_train_standardized, y_train)

predictions_standardized = {
    'Logistic Regression': logreg_standardized.predict(X_test_standardized),
    'Decision Tree': decision_tree_standardized.predict(X_test_standardized),
    'Random Forest': random_forest_standardized.predict(X_test_standardized),
    'XGBoost': xgb_model_standardized.predict(X_test_standardized)
}

for name, y_pred_standardized in predictions_standardized.items():
    print(f"Classification Report for {name} with Standardized Data:")
    print(classification_report(y_test, y_pred_standardized))
    print("-" * 50)
```

### confusion matrix

```python
fig, axes = plt.subplots(1, len(predictions_standardized), figsize=(20, 5))
fig.suptitle('Confusion Matrices for Standardized Data')
for ax, (name, y_pred) in zip(axes.flatten(), predictions_standardized.items()):
    disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax, cmap='Blues')
    disp.ax_.set_title(name)
plt.tight_layout()
plt.show()
```

### cross validation

```python
from sklearn.model_selection import cross_val_score

models = {
    'Logistic Regression': logreg_standardized,
    'Decision Tree': decision_tree_standardized,
    'Random Forest': random_forest_standardized,
    'XGBoost': xgb_model_standardized
}

scoring_metrics = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

val_score_data = []

for metric in scoring_metrics:
    for name, model in models.items():
        scores = cross_val_score(model, X_train_standardized, y_train, scoring=metric, cv=10, n_jobs=-1)
        val_score_data.append({
            'Model': name,
            'Metric': metric,
            'Mean Score': np.mean(scores),
            'Standard Deviation': np.std(scores)
        })
val_score_df = pd.DataFrame(val_score_data)

val_score_df_pivot = val_score_df.pivot(index='Model', columns='Metric', values=['Mean Score', 'Standard Deviation'])

print(val_score_df_pivot)
```

**Classification Reports:** \\
Looking at the classification reports for both normalization and standardization, we can see that the performance metrics (precision, recall, f1-score, and accuracy) are quite similar for most models across both types of data. \\
In general: \\
Logistic Regression: The performance is similar for both types of data.
Decision Tree, Random Forest, and XGBoost: The performance is consistent across both types of data. \\

**Cross-Validation Results:** \\
The cross-validation results also show similar mean scores for accuracy, f1_macro, precision_macro, and recall_macro across both normalization and standardization. The standard deviation of these metrics is also comparable for both types of data. \\

**Conclusion:**
Based on the classification reports and cross-validation results, both normalization and standardization perform similarly well for the given models and dataset. There is no clear indication that one preprocessing technique is significantly better than the other for this task.
